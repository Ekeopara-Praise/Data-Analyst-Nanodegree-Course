## Introduction

The phrase, Man's best friend is a common phrase used to describe domestic dogs, referring to their millennia-long history of close relations, loyalty, friendship, and companionship with humans [Wikipedia](https://en.wikipedia.org/wiki/Man%27s_best_friend). Dogs have always been the pet of various homes and the joy and friendship they bring can never be over emphasized. There are various dog stages and the choice of one is dependent on personal views. However, are there interesting insights we can see from dog various collections? And what are the human views of specific dogs?

In this data wrangling project, which is a part of the ALT-X Udacity Project, we will see some insights from the wrangling process. This process involved data gathering, assessing the data quality and cleaning.  

## About the Data

There were three datasets from different sources which were gathered together for this analysis. They include;

1. **The WeRateDogs Twitter** - This is a Twitter archive that contains basic tweet data for all 5000+ of their tweets on the  WeRateDogs account.
2. **Image Predictions data** - Contains various image predictions of dogs for these twitter acount.
3. **Additional data from the Twitter API** - Additional data including tweet ID, retweet count, and favorite count etc.

## Insights and Visualizations

After carrying out the three data wrangling processes (gathering, assessing and cleaning), the datasets were merged together and used for visulisations. These insights can be summarized by answering these basic questions?

> 1. **What are the common names of dogs?**
2. **Which is the highest dog rating present in the tweets?**
3. **Which Dog's stage has the highest ratings?**
4. **Which Dog's stage has the highest Favourite Count?**

**What are the common names of dogs?**

> #### Observation
While most persons did not specify the name of their dogs, it appears that majority of the dogs here are named Oliver, tucker, bella, cooper, clarence, winston lucy etc.

 **Which is the highest dog rating present in the tweets?**

> #### Observation

**Which Dog stage has the highest ratings?**

> #### Observation

**Which Dog's stage has the highest Favourite Count?**

> #### Observation
